```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import cv2 as cv
```

```python
def imshow(img, st='Untitled Image'):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(st)
    plt.show()
```

```python
sudoku = cv.imread('sudoku.jpg')
sudo = sudoku.copy()
sudoku_gray = cv.cvtColor(sudoku, cv.COLOR_BGR2GRAY)
```

```python
imshow(sudoku)
imshow(sudoku_gray)
```

```python
ret, sud = cv.threshold(sudoku_gray, 100, 255, cv.THRESH_BINARY)
imshow(sud)
```

```python
cnt, heir = cv.findContours(sud, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
cv.drawContours(sudo, cnt, -1, (255, 0, 0), 3, cv.LINE_AA)
imshow(sudo)
suddd = sudoku.copy()
full = np.full_like(suddd, 255)
# cv.drawContours(suddd, cnt, -1, (255, 0, 0), -1, cv.LINE_AA)
# cv.drawContours(full, cnt, -1, (255, 0, 0), 3, cv.LINE_AA)
for c in cnt:
    if cv.contourArea(c) < 10:
        cv.drawContours(suddd, cnt, -1, (255, 0, 0), -1, cv.LINE_AA)
imshow(suddd)
imshow(full)
```
