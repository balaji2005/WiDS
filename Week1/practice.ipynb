```python
img_gray = cv.imread('image.jpg')
```

```python
def imshow(img, st='Untitled Image'):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(st)
    plt.show()
```

```python
ret, img_thresh = cv.threshold(img_gray, 100, 255, cv.THRESH_BINARY_INV)
imshow(img_gray, 'Rose')
```

```python
assign_copyg = assign_gray.copy()
ret, assign_copyg = cv.threshold(assign_copyg, 100, 255, cv.THRESH_BINARY_INV)
imshow(assign_copyg)

contours = cv.findContours(assign_copyg, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)
contours = max(contours, key=cv.contourArea)

for eps in np.linspace(0.001, 0.1, 20):
	# approximate the contour
	peri = cv.arcLength(contours, True)
	contourB = cv.approxPolyDP(contours, eps * peri, True)

assign_copy = assign.copy()
assign_copy = cv.drawContours(assign_copy, [contourB], -1, (255, 0, 0), 1, cv.LINE_AA)
imshow(assign_copy)
```
